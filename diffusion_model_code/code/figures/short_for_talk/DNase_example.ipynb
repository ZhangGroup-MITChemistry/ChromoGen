```python
import matplotlib.pyplot as plt
plt.style.use('/home/gridsan/gschuette/universal/matplotlib/plot_style.txt')

import numpy as np
```

```python
def new_y(last,mi,ma):

    dy = ma - mi
    std = dy / 20 # the rough size of the window over which new y values are allows to move

    mean = min(last,ma-std)
    mean = max(mean,mi+std)

    new = np.random.randn()*std + mean
    new = min(new,ma)
    new = max(new,mi)

    return new
    
    
    

def y_with_weight(npoints,minimum=0,maximum=1):

    n, mi, ma = npoints, minimum, maximum 
    dy = ma - mi # Size of range
    
    y = np.empty(n)

    y[0] = np.random.rand() * dy + mi 

    for i in range(1,n):
        y[i] = new_y(y[i-1],mi,ma)
        

    return y

    
    
```

```python
npoints = 1000
x = np.arange(npoints) 
y = y_with_weight(npoints)

fig = plt.figure(figsize=(6,.5))
ax = fig.add_subplot(111)
ax.plot(x,y)

ax.set_xlim([0,npoints])
ax.set_ylim([-.05,1.05])
ax.set_xticks([])
ax.set_yticks([])

#ax.
```
