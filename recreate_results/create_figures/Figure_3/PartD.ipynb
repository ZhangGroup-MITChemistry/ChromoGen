The following are parameters for the distance -> contact probability conversion, i.e.
$$
p_{ij}(r_{ij}) = 
\begin{cases}
    \frac{\left[ 1 + \tanh\left( \sigma\left( r_c - r_{ij} \right) \right) \right]}{2} & r_{ij} < r_c \\
    \left( \frac{r_c}{r_{ij}} \right)^{\texttt{decay\_rate}} & \text{otherwise}
\end{cases}
$$
where $p_{ij}$ and $r_{ij}$ are the contact probability and spatial separation between monomers $i$ and $j$, respectively, $r_c$ is the cutoff distance, $\sigma$ determines the relative sharpness of the probability decay at short distances, and $\texttt{decay\_rate}$ determines the probability decay at long distances. 

```python
r_c = 1.5 
sigma = 3.72
decay_rate = 3.45

assert isinstance(r_c,float)
assert isinstance(sigma,float)
assert isinstance(decay_rate,float)
```

```python
def load_region(
    chrom,start_idx,scales_rescales=[(1.,0.),(5.,8.)],
    *,d=cgen_conformations_dir,config_ds=config_ds,
    hic_dataset=hic_dataset,use_gpu=use_gpu,
    high_precision=high_precision,
    resolution=resolution
):
    #####
    # Load data
    
    # Load/combine conformations inferred with ChromoGen at each set of guidance parameters
    cgen_coords = []
    for cond_scale, rescaled_phi in scales_rescales:
        fp = cgen_conformations_dir / \
        f'chrom_{chrom}/start_idx_{start_idx//1000}_cond_scale_{cond_scale}_rescaled_phi_{rescaled_phi}.pt'
        cgen_coords.append(Conformations(fp,drop_invalid_conformations=True))
    cgen_coords = cgen_coords[0].cat(cgen_coords[1:])

    # Load the Dip-C data
    dipc_coords = Conformations(config_ds.fetch_specific_coords(str(chrom),start_idx)[-1])

    # Experimental Hi-C
    stop = start_idx + dipc_coords.num_beads * resolution
    exp_hic = hic_dataset.fetch(str(chrom), start_idx, stop)

    ####
    # Set dtype, device
    
    # GPU?
    if use_gpu:
        cgen_coords.cuda_()
    else:
        # Just to be extra sure
        cgen_coords.cpu_()

    # Double or float?
    if high_precision:
        cgen_coords.double_()
    else:
        # Just to be extra sure
        cgen_coords.float_()

    # Just to be extra sure things are on the same device, etc. 
    d,dt = cgen_coords.device, cgen_coords.dtype
    dipc_coords.to_(dtype=dt,device=d)
    exp_hic.to_(dtype=dt,device=d)
        
    return cgen_coords, dipc_coords, exp_hic
```

```python
def process_coords(coords,r_c=r_c,sigma=sigma,decay_rate=decay_rate):
    dists = coords.distances
    return dists.median, dists.compute_hic(r_c, sigma, decay_rate)
```

```python
def compute_pcc(x,y,x_log=False, y_log=False,drop_nan=False):
    # x & y are either a Distances or HiCMap instance
    x = x.values.squeeze()
    y = y.values.squeeze()

    assert x.ndim==y.ndim==2, 'Should only pass one sample for x & y.'

    # Prepare for torch.corrcoef
    n = x.shape[-2]
    i,j = torch.triu_indices(n,n,1) # Ignore diagonal b/c it's all zeros (dists)/ones (probs)
    data = torch.stack(
        [
            x[i,j],
            y[i,j]
        ],
        dim=0
    )

    # Ignore NaN values
    if drop_nan:
        data = data[:, data.isfinite().all(0)]

    if data.shape[-1] == 0:
        return torch.nan

    # Take the log of neither, one, or both sets of data
    if x_log:
        if y_log:
            data.log_()
        else:
            data[0,:].log_()
    elif y_log:
        data[1,:].log_()

    # Compute & return the pearson correlation coefficient
    return torch.corrcoef(data)[0,1].cpu()
```

```python
def full_pipeline(chrom, start):
    ###
    # Preparation
    
    # Load
    cgen_coords, dipc_coords, exp_hic = load_region(chrom, start)

    # Get median distances & infer Hi-C maps
    cgen_median, cgen_hic = process_coords(cgen_coords)
    dipc_median, dipc_hic = process_coords(dipc_coords)

    ###
    # Correlation between:
    
    # Median distances
    # For median distances, skip regions with < 80 Dip-C 
    # conformations to ensure agreement isn't due to statistical noise.
    # The NaN values are ignored downstream
    if len(dipc_coords) < 80:
        md_md = torch.nan
    else:
        # The original script used the following PCC implementation, which slightly affects
        # the appearance of this plot. So, don't use the compute_pcc function. 
        #md_md = compute_pcc(cgen_median, dipc_median)
        n = cgen_median.num_beads
        i,j = torch.triu_indices(n,n,1)
        a = cgen_median.values[0,i,j]
        b = dipc_median.values[0,i,j]
        md_md = ( ( (a*b).mean() - a.mean()*b.mean() ) / (a.std() * b.std()) ).cpu()

    # Median distance & log probability
    # Hi-C maps contain NaN's in many regions, so ignore them
    # for a more complete picture
    md_lp_cgen = compute_pcc(cgen_median, exp_hic, False, True, True)
    md_lp_dipc = compute_pcc(dipc_median, exp_hic, False, True, True)

    # Hi-C vs Hi-C 
    lp_lp_cgen = compute_pcc(cgen_hic, exp_hic, False, False, True)
    lp_lp_dipc = compute_pcc(dipc_hic, exp_hic, False, False, True)
    
    return md_md, md_lp_cgen, md_lp_dipc, lp_lp_cgen, lp_lp_dipc
```

## A quick illustration

Load a set of conformations, compute mean distance map & infer the Hi-C probabilities, and plot them. 

```python
chrom = 11
start_idx = 129_720_000

####
# Load data
cgen_coords, dipc_coords, exp_hic = load_region(chrom,start_idx)

####
# Prep data 

# Coordinates -> Distances
cgen_dists = cgen_coords.distances
dipc_dists = dipc_coords.distances
del cgen_coords, dipc_coords

# Distances -> Median distances
cgen_med_dists = cgen_dists.median
dipc_med_dists = dipc_dists.median

# Distances -> Hi-C probabilities
# NOTE: Could also do this straight from coords, but then distances 
# would have to be computed twice
cgen_hic = cgen_dists.compute_hic(r_c, sigma, decay_rate)
dipc_hic = dipc_dists.compute_hic(r_c, sigma, decay_rate)
del cgen_dists, dipc_dists

# Normalize to nearest-neighbor contact probabilities in the 
# experimental Hi-C, which improves visualization/accounts for the
# fact that nearest neighbors (almost) always interact in a polymer
# model, but not in experimental Hi-C. 
# Note that we DON'T do this elsewhere in this script, but it doesn't
# actually affect correlation coefficients anyway
exp_hic.normalize_as_(cgen_hic)
# Same as: exp_hic.normalize_(cgen_hic.diagonal_mean(offset=1))

# Hi-C vmin, for consistency
v1,v2,v3 = cgen_hic.values, dipc_hic.values, exp_hic.values
vmin = min(v1[v1>0].min(),v2[v2>0].min(),v3[v3>0].min())

####
# Make figure

fig, axes = plt.subplots(nrows=1,ncols=3,layout='constrained',figsize=(7,3))

*_,cbar = exp_hic.plot(fig=fig,ax=axes[0],cbar_orientation='horizontal',vmin=vmin)
cbar.set_label('Experimental Hi-C',wrap=True)

*_,cbar = dipc_hic.plot_with(cgen_hic,fig=fig,ax=axes[1],cbar_orientation='horizontal',vmin=vmin)
cbar.set_label('Distance-inferred Hi-C',wrap=True)
axes[1].set_ylabel('')

*_,cbar = dipc_med_dists.plot_with(cgen_med_dists,fig=fig,ax=axes[2],cbar_orientation='horizontal')
cbar.set_label('Median distances (reduced units)',wrap=True)
axes[2].set_ylabel('')

del cgen_med_dists, dipc_med_dists, cgen_hic, dipc_hic, v1, v2, v3, vmin
```

## Compute statistics

```python
data_files = [
    temp_data_dir / f'{x}.pt'
    for x in ['dist_v_dist','dist_v_prob_cgen','dist_v_prob_dipc','prob_v_prob_cgen','prob_v_prob_dipc']
]

if False:#all(f.exists() for f in data_files):
    
    dist_v_dist = torch.load(data_files[0])
    dist_v_prob_cgen = torch.load(data_files[1])
    dist_v_prob_dipc = torch.load(data_files[2])
    prob_v_prob_cgen = torch.load(data_files[3])
    prob_v_prob_dipc = torch.load(data_files[4])

else:

    dist_v_dist = []
    dist_v_prob_cgen = []
    dist_v_prob_dipc = []
    prob_v_prob_cgen = []
    prob_v_prob_dipc = []

    with ThreadPoolExecutor(max_workers=max_workers) as executor:
        futures = []
        # loop over ONE set of the conditioning parameters so that we perform each set of computations exactly once
        files = list(cgen_conformations_dir.rglob('*1.0_rescaled_phi_0.0.pt'))
        files.sort()
        for f in tqdm(files,desc='Initiating tasks',total=2658,initial=0):
            start_idx = int(f.name.split('_')[2]) * 1000
            chrom = f.parent.name.split('_')[-1]
    
            # Excluding chromosome X
            if chrom == 'X':
                continue
    
            futures.append(executor.submit(full_pipeline, chrom, start_idx))
    
        for future in tqdm(futures,desc='Processing data'):
            a,b,c,d,e = future.result()
            dist_v_dist.append(a)
            dist_v_prob_cgen.append(b)
            dist_v_prob_dipc.append(c)
            prob_v_prob_cgen.append(d)
            prob_v_prob_dipc.append(e)

    dist_v_dist = torch.tensor(dist_v_dist)
    dist_v_prob_cgen = torch.tensor(dist_v_prob_cgen)
    dist_v_prob_dipc = torch.tensor(dist_v_prob_dipc)
    prob_v_prob_cgen = torch.tensor(prob_v_prob_cgen)
    prob_v_prob_dipc = torch.tensor(prob_v_prob_dipc)

    torch.save(dist_v_dist,data_files[0])
    torch.save(dist_v_prob_cgen,data_files[1])
    torch.save(dist_v_prob_dipc,data_files[2])
    torch.save(prob_v_prob_cgen,data_files[3])
    torch.save(prob_v_prob_dipc,data_files[4])
        

```

```python
def plot_hist(
    data1,
    data2,
    tag1,
    tag2,
    nbins=25,
    fig=None,
    ax=None,
    alpha=0.5,
    bins=None
):

    # Convert to numpy to improve speed in plt 
    data1 = data1[data1.isfinite()].cpu().numpy()

    # Handle data2 not None vs None cases
    if data2 is None:
        if bins is None:
            minn = data1.min()
            maxx = data1.max()
    else:
        data2 = data2[data2.isfinite()].cpu().numpy()
        if bins is None:
            minn = min(data1.min(),data2.min())
            maxx = max(data1.max(),data2.max())

    # Create consistent bins for each data type
    if bins is None:
        delta = (maxx-minn)/nbins
        bins = np.arange(minn,maxx+delta/2,delta)

    # Create the figure, if necessary
    if ax is None:
        fig = plt.figure()
        ax = fig.add_subplot(111)
        
    # Create the histogram
    ax.hist(data1,alpha=alpha,bins=bins,label=tag1,edgecolor='none',density=True)
    if data2 is not None:
        ax.hist(data2,alpha=alpha,bins=bins,label=tag2,edgecolor='none',density=True)

    return fig, ax
```
