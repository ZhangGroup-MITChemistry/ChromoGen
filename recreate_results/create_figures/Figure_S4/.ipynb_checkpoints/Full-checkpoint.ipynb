## Prep

Import packages

```python
from ChromoGen import Conformations
from ChromoGen.data_utils import HiCDataset
import torch
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib.patheffects as pe
plt.style.use('../support/plot_style.txt')
```

Settings

```python
mcool_fp = '../../outside_data/hic/GM12878_hg19.mcool'
cgen_conformation_dirs = [
    '../../downloaded_data/conformations/ChromoGen/genome_wide/GM12878/',
    '../../downloaded_data/conformations/ChromoGen/specific_regions/GM12878/'
]
pdf_dir = './pdfs'

resolution = 20_000

regions_to_analyze = {
    '2':[24_960_000],
    '4':[28_100_000],
    '6':[48_040_000],
    '11':[129_720_000],
    '18':[5_200_000],
    'X':[121_020_000]
}

guidance_params = [(1.,0.),(5.,8.)]

# distance -> Hi-C probability settings
r_c = 1.5
sigma = 3.72
decay_rate = 3.45

# Insulation score computation's window size
window_size = 7
```

Initialize objects

```python
hic_dataset = HiCDataset(mcool_fp)
cgen_conformation_dirs = [Path(d) for d in cgen_conformation_dirs]

pdf_dir = Path(pdf_dir)
pdf_dir.mkdir(exist_ok=True,parents=True)
```

## Support functions

```python
def pipeline(
    chrom,
    start_idx,
    hic_dataset=hic_dataset,
    coord_dirs=cgen_conformation_dirs,
    guidance_params=guidance_params,
    resolution=resolution,
    r_c=r_c,
    sigma=sigma,
    decay_rate=decay_rate,
    window_size=window_size
):
    ############################
    # Load/compute all Hi-C data
    
    ####
    # ChromoGen-inferred Hi-C

    # Load coordinates for this region
    cgen_coords = []
    for cs, rp in guidance_params:
        fn = f'chrom_{chrom}/start_idx_{start_idx//1000}_cond_scale_{cs}_rescaled_phi_{rp}.pt'
        for k,d in enumerate(coord_dirs):
            if (f:= d/fn).exists():
                cgen_coords.append(Conformations(
                    f,
                    drop_invalid_conformations=True
                ))
                break
            elif k == len(coord_dirs)-1:
                raise Exception(
                    f'No generated structures found for chromosome {chrom}: {round(start_idx/1_000_000,2)}-<end> '
                    f'Mb using cond_scale {cs} and rescaled_phi {rp}. {d/fn}'
                )
    cgen_coords = cgen_coords[0].extend(cgen_coords[1:]).double()

    # Compute Hi-C 
    cgen_hic = cgen_coords.compute_hic(r_c, sigma, decay_rate)
    del cgen_coords
    
    ####
    # Experimental Hi-C 

    # Load
    stop = start_idx + cgen_hic.nbins * resolution
    exp_hic = hic_dataset.fetch(chrom,start_idx,stop).to_(cgen_hic.dtype)

    # Normalize to the ChromoGen-inferred map to improved visualization. 
    # This simply scales the experimental map by a constant such that the average value 
    # on the first off diagonal is equal in both maps. 
    # Note that scaling by a constant doesn't affect the resulting pearson correlation coefficient
    # of the insulation scores. 
    exp_hic.normalize_as_(cgen_hic)

    ############################
    # Compute the insulation scores, correlations

    # Insulation scores
    cgen_IS = cgen_hic.compute_insulation(window_size=window_size)
    exp_IS = exp_hic.compute_insulation(window_size=window_size)

    ####
    # Correlation

    # Combine data for use in torch.corrcoef
    v = torch.stack([cgen_IS, exp_IS], dim=0)

    # Drop the NaN values at the edges of the map (where the window goes out-of-range)
    v = v[:,v.isfinite().all(0)]

    # Compute correlation
    corr = float(torch.corrcoef(v)[0,1])

    ############################
    # Return results
    return cgen_hic, exp_hic, cgen_IS, exp_IS, corr
```

```python
def process_plot_region(chrom, start_idx, resolution=resolution, **kwargs):

    ########################
    # Load/prep/compute data
    cgen_hic, exp_hic, cgen_IS, exp_IS, corr = pipeline(chrom,start_idx,resolution=resolution,**kwargs)

    ########################
    # Make the plots

    ########
    # Initialize
    fig,axes = plt.subplots(ncols=2,nrows=1,figsize=(3,2),layout='constrained')

    # Set the figure title
    stop = start_idx + resolution * len(cgen_IS) 
    fig.suptitle(
        f'Chromosome {chrom}: {round(start_idx/1e6,2)}-{round(stop/1e6,2)} Mb, '
        '$\\rho = 'f'{round(corr,2)}$'
    )

    ########
    # Hi-C panel
    ax = axes[0]
    
    # Plot the Hi-C
    *_, cbar = exp_hic.plot_with(cgen_hic, fig=fig, ax=ax, cbar_orientation='horizontal')

    # Remove the cbar label & ylabel
    cbar.set_label('')
    ax.set_ylabel('')

    # Label upper/lower triangles
    n = exp_hic.nbins
    ax.text(.995*n-1,.005*n,'Expt. Hi-C',horizontalalignment='right',verticalalignment='top',
            color='black',path_effects=[pe.withStroke(linewidth=1, foreground="white")],fontsize='small')
    
    ax.text(.005*n,.995*n-1,'ChromoGen',horizontalalignment='left',verticalalignment='baseline',
            color='black',path_effects=[pe.withStroke(linewidth=1, foreground="white")],fontsize='small')
    
    ########
    # Insulation score panel
    ax = axes[1]

    ####
    # create an axis that we'll leave blank to account for the space taken up by the colorbar 
    # in the Hi-C plot
    #divider = make_axes_locatable(ax)
    #cax = divider.append_axes("bottom", size="5%", pad=0.05)

    ####
    # Plot the data
    
    # Have x reflect the genomic index, centering bins
    x = torch.arange(start_idx, stop + resolution/2, resolution)
    x = (x[1:] + x[:-1]) / 2e6 # e6 converts to Mb

    # Plot the insulation scores
    ax.plot(x,cgen_IS,label='ChromoGen', linewidth=1.5, alpha=.85)
    ax.plot(x,exp_IS,label='Expt. Hi-C', linewidth=1.5, alpha=.85)

    ####
    # Format the axes
    
    # Set x-axis limits
    xlim = [start_idx / 1e6, stop / 1e6]
    ax.set_xlim(xlim)

    # Choose & set y-axis limits
    data_min = min( float(cgen_IS[cgen_IS.isfinite()].min()), float(exp_IS[exp_IS.isfinite()].min()) )
    data_max = max( float(cgen_IS[cgen_IS.isfinite()].max()), float(exp_IS[exp_IS.isfinite()].max()) )
    y_center = (data_min + data_max) / 2
    half_range = (data_max - data_min) * .5 
    ylim = [y_center - half_range*1.9, y_center + half_range*1.05] # Pad slightly on top, enough on bottom to fit legend
    ax.set_ylim(ylim)

    # Place the yticks & label on the right side
    ax.yaxis.tick_right()
    ax.yaxis.set_label_position("right")

    # Set labels
    ax.set_xlabel('Genomic index (Mb)')
    ax.set_ylabel('Insulation score',rotation=270,va='baseline')
    
    ####
    # Miscellaneous

    # Add in the legend
    ax.legend(frameon=False, loc='lower right')

    # Force square plot
    ax.set_aspect((xlim[1]-xlim[0])/(ylim[1]-ylim[0]))

    # Remove the cax
    #cax.remove()

    return fig
```

```python
for chrom, starts in regions_to_analyze.items():
    for start_idx in starts:
        fig = process_plot_region(chrom,start_idx)
        fig.savefig(pdf_dir/f'chrom_{chrom}_{start_idx//1000}.pdf',transparent=True)
```
