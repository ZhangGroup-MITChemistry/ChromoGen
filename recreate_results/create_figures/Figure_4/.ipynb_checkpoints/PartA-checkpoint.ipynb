```python
import pyBigWig
import numpy as np
from pathlib import Path
from ChromoGen import Conformations
from ChromoGen.data_utils import HiCDataset
import matplotlib.pyplot as plt
import matplotlib.patheffects as pe
plt.style.use('../support/plot_style.txt')
```

```python
gm_bigWig_fp = '../../outside_data/sequence_data/DNase_seq/hg19/GM12878.bigWig'
imr_bigWig_fp = '../../outside_data/sequence_data/DNase_seq/hg19/IMR90.bigWig'

gm_mcool_fp = '../../outside_data/hic/GM12878_hg19.mcool'
imr_mcool_fp = '../../outside_data/hic/IMR90_hg19.mcool'

cgen_gm_conformations_dir = '../../downloaded_data/conformations/ChromoGen/specific_regions/GM12878/'
cgen_imr_conformations_dir = '../../downloaded_data/conformations/ChromoGen/specific_regions/IMR90/'

temp_data_dir = './temp_data/'
pdf_dir = './pdfs/'

region_length=1_280_000
plotting_resolution = 1_000 # for DNase-seq data
resolution = 20_000 # For distance/Hi-C plots

chrom_start_idx = {
    '6':[6_000_000, 131_860_000],
    '21':[28_500_000]
}

dist_vmins = {
    '6':[100,50],
    '21':[150]
}

GM_on = 'right' 
IMR_on = 'left'
```

## Prepare

### Verify filepaths, intialize data objects

```python
assert isinstance(plotting_resolution,int) and plotting_resolution > 0, 'plotting_resolution must be a positive integer'
assert isinstance(region_length,int) and region_length > 0, 'region_length must be a positive integer'
assert region_length >= plotting_resolution, 'plotting_resolution cannot be larger than region_length'
assert region_length % plotting_resolution == 0, 'region_length must be divisible by plotting_resolution'

assert all(x in ['left','right'] for x in [GM_on,IMR_on]), 'GM_on and IMR_on must be either "left" or "right".'
orientations = {
    'GM12878':GM_on,
    'IMR-90':IMR_on
}

bigWigs = {
    'GM12878':pyBigWig.open(gm_bigWig_fp),
    'IMR-90':pyBigWig.open(imr_bigWig_fp)
}

hic_datasets = {
    'GM12878':HiCDataset(gm_mcool_fp,resolution=resolution),
    'IMR-90':HiCDataset(imr_mcool_fp,resolution=resolution)
}

for chrom,start_indices in chrom_start_idx.copy().items():
    
    # format chrom name since pyBigWig is finicky about it & I'm used to the greater flexibility 
    # regarding chromosome naming conventions provided by my classes
    chrom1 = f"chr{str(chrom).replace('chr','')}"
    
    if isinstance(start_indices, int):
        si = [start_indices]
    else:
        si = start_indices
    assert isinstance(si, list), 'Start indices inside chrom_start_idx should be placed inside a list'
    assert si, 'Cannot have chromosome in chrom_start_idx without any start indices defined'
    assert all(isinstance(x,int) and x >= 0 for x in si), 'Start indices must be integers >= 0'
    
    l = set(bw.chroms(chrom1) for bw in bigWigs.values())
    assert len(l) == 1, (
        'BigWig files appear to correspond to different assemblies! '
        f'Lengths in chromosome {chrom} are {l}'
    )
    l = l.pop()
    
    l1 = set(hic_dataset.clr.chromsizes[chrom1[3:]] for hic_dataset in hic_datasets.values())
    assert len(l1) == 1, (
        'mcool files appear to correspond to different assemblies! '
        f'Lengths in chromosome {chrom} are {l1} for the files'
    )
    l1 = l1.pop()
    assert l==l1, (
        'mcool and bigWig files appear to correspond to different assemblies! '
        f'Chromosome {chrom} is length {l:,} in bigWig files and {l1:,} in mcool files.'
    )
    
    for x in start_indices:
        assert x + region_length <= l, (
            f'Start position {x:,} with region length {region_length:,} is out of bounds '
            f'in chromosome {chrom}, which has length {l:,}.'
        )
    
    # Use the formatted chromosome name
    chrom_start_idx[chrom1] = chrom_start_idx.pop(chrom)

    # Check/format the vmins. Assume the chromosome naming convention is the same as in chrom_start_idx
    if chrom not in dist_vmins:
        dist_vmins[chrom1] = [None] * len(start_indices)
    else:
        vmins = dist_vmins.pop(chrom)
        if vmins is None:
            vmins = [None] * len(start_indices)
        elif isinstance(vmins,(int,float)):
            vmins = [vmins] * len(start_indices)
        else:
            assert isinstance(vmins,list), 'values in dist_vmins dict should be None, int/float, or list of these.'
            if len(vmins) != len(start_indices):
                assert len(vmins)==1, (
                    'list values in dist_vmins should have length 1 (apply to all) '
                    'or same length as the corresponding value in the chrom_start_idx dict.'
                )
                vmins = vmins * len(start_indices)
        dist_vmins[chrom1] = vmins

assert len(dist_vmins)==len(chrom_start_idx), \
'There are chromosomes defined in dist_vmins but not chrom_start_idx. Only the opposite is ok'

cgen_imr_conformations_dir = Path(cgen_imr_conformations_dir)
assert cgen_imr_conformations_dir.exists()

cgen_gm_conformations_dir = Path(cgen_gm_conformations_dir)
assert cgen_gm_conformations_dir.exists()

data_dirs = {
    'GM12878':cgen_gm_conformations_dir,
    'IMR-90':cgen_imr_conformations_dir
}

temp_data_dir = Path(temp_data_dir)
temp_data_dir.mkdir(exist_ok=True,parents=True)

pdf_dir = Path(pdf_dir)
pdf_dir.mkdir(exist_ok=True,parents=True)
```

### Support functions for plotting

#### DNase-seq plots

```python
def plot_dnase(
    data,
    start_idx,
    resolution,
    orientation,
    data_max = None,
    pre_coarsened=False
):

    ########################################
    # Prep 

    ####
    # Coarsen to the desired resolution, if needed
    if not pre_coarsened:
        data = data.reshape(-1,resolution).mean(1)

    ####
    # Get the genomic indices associated with each value
    # we're plotting. Use the center of each bin. 
    start = start_idx + resolution // 2
    stop = start + len(data) * resolution
    seq = np.arange(start, stop, resolution)

    # Divide by 1,000,000 to get Mb for cleaner labels
    seq = seq / 1e6

    # Subtract the resolution/2 so start and stop can be reused later
    start-= resolution//2
    stop-= resolution//2

    ####
    # Get axis limits, ticks, tick labels. 
    # We'll account for left vs right orientation later
    if data_max is None:
        data_max = np.nanmax(data)

    xmax = data_max*1.075
    xlim = [data_max-xmax,xmax]
    ''' Numbers nearly bump into each other if we do it this way...
    xticks = [0, int(xmax) if xmax%1 < .5 else round(int(xmax)+.5,1)]
    '''
    xticks = [0, 2.5 if xmax > 2 else 1] 
    #xticklabels = [f'${xt}$' for xt in xticks]
    xticklabels = [f'{xt}' for xt in xticks]
    ylim = yticks = [ # Reverse order because that's how distance map y-axes are
        round( (start_idx + len(data) * resolution)/1e6, 2),
        round(start_idx/1e6,2)
    ]
    #yticklabels = [f'${yt}$' for yt in yticks]
    yticklabels = [f'{yt}' for yt in yticks]
    
    ########################################
    # Plot

    ####
    # Initialize
    fig,ax = plt.subplots(ncols=1,nrows=1,figsize=(.7,2))

    ####
    # Plot data, setting sequence on y-axis to create the 'rotated' plot used in the paper
    ax.plot(data,seq)

    ####
    # Format

    # Account for left vs right
    if orientation == 'left':
        xlim = list(reversed(xlim))
        xticks = list(reversed(xticks))
        xticklabels = list(reversed(xticklabels))
        rot = 90 # Rotation of ylabel
    else:
        ax.yaxis.tick_right()
        ax.yaxis.set_label_position("right")
        rot = 270

    # Set axis limits
    ax.set_ylim(ylim)
    ax.set_xlim(xlim)
    
    # Set ticks, ticklabels
    ax.set_xticks(xticks, labels = xticklabels)
    ax.set_yticks(yticks, labels = yticklabels, rotation=rot)

    # Set 'title' (ylabel)
    ax.set_ylabel('Genomic index (Mb)', rotation=rot)

    # Fix aspect for consistency
    ax.set_aspect(17/3 * ( (max(xlim)-min(xlim))/(max(ylim)-min(ylim)) ))

    return fig, ax
    
    # Set axis limits, ticks, tick labels, other labels 
    # Set y-axis limits. Go all the way sequence extremes even though we binned the data
    ax.set_ylim(ylim)

    if orientation == 'left':
        xlim = list(reversed(xlim))
        ax.set_yticks()

    if orientation == 'left':
        reversed
        ax.set_yticks(ylim,ylim,rotation='vertical')
        ax.set_ylabel('Genomic index (Mb)')
        ax.set_xlim(xlim)
        
    elif orientation == 'right':
        ax.yaxis.set_label_position("right")
        ax.yaxis.tick_right()
        ax.set_yticks(ylim,ylim,rotation=270)
        ax.set_ylabel('Genomic index (Mb)',rotation=270,x=1, y=.5,va='center',ha='center')
        ax.set_xlim(xlim)
    
    return fig, ax

def make_plots(
    chrom,
    start_idx,
    cell_type=None,
    bigWigs = bigWigs,
    region_length = region_length,
    orientations = orientations,
    resolution = plotting_resolution,
    
):

    cell_types = list(bigWigs) if cell_type is None else [cell_type]
    
    # Do it this weird way to ensure axis limits are the same
    args = {}
    data_max = 0.
    for cell_type in cell_types:
        args[cell_type] = {}

        # Left or right?
        args[cell_type]['orientation'] = orientations[cell_type]

        ####
        # Load/format DNase-seq data

        # Get bigwig & raw data
        bw = bigWigs[cell_type]
        data = np.array(bw.values(chrom, start_idx, start_idx + region_length))

        # Coarsen to the desired resolution
        data = data.reshape(-1,resolution).mean(1)

        # Get max value & compare to max from other cell types
        data_max = max(data_max, float(np.nanmax(data)) )

        # Place in the args dict
        args[cell_type]['data'] = data

    figs_axes = {}
    for cell_type, kwargs in args.items():
        figs_axes[cell_type] = plot_dnase(
            start_idx=start_idx, 
            resolution=resolution, 
            data_max=data_max,
            pre_coarsened=True,
            **kwargs
        )

    return figs_axes
    
```

#### Distance/Hi-C plots

```python
def load_data(
    chrom, 
    start_idx, 
    hic_datasets=hic_datasets, 
    cgen_imr_dir=cgen_imr_conformations_dir,
    cgen_gm_dir=cgen_gm_conformations_dir,
    resolution = resolution,
    interp_nans = False
):
    si = start_idx
    si1 = si // 1000
    r = resolution

    med_dists = {}
    exp_hic = {}
    for cell_type, d in [('GM12878',cgen_gm_dir),('IMR-90',cgen_imr_dir)]:
        # median distances
        f1 = d / f'chrom_{chrom}/start_idx_{si1}_cond_scale_1.0_rescaled_phi_0.0.pt'
        f2 = d / f'chrom_{chrom}/start_idx_{si1}_cond_scale_5.0_rescaled_phi_8.0.pt'
        coords = Conformations(f1,drop_invalid_conformations=True).cat(
            Conformations(f2,drop_invalid_conformations=True)
        )
        med_dists[cell_type] = d = coords.distances.median

        # to nm
        d._values*= 100

        n = d.num_beads
        exp_hic[cell_type] = hic_datasets[cell_type].fetch(chrom, si, si+r*n, interp_nans=interp_nans)

    return med_dists, exp_hic
    


def plot_hic_distance(
    chrom,
    start_idx,
    cell_type_upper='GM12878',
    cell_type_lower='IMR-90',
    cbar_orientation='horizontal',
    dist_vmin=None,
    dist_vmax=None,
    interp_nans=False
):

    ####
    # Prep 
    
    # Format for use with our mcool files... which don't follow standard convention...
    chrom1 = chrom.replace('chr','')

    # Get data
    med_dists, exp_hic = load_data(chrom1, start_idx,interp_nans=interp_nans)

    ####
    # Plot

    # Initialize plot
    fig,(hic_ax,dist_ax)=plt.subplots(ncols=2, nrows=1, figsize=(7.08*.5*2/3,6.692*.25),layout="constrained")

    # Hi-C
    exp_hic1 = exp_hic[cell_type_upper]
    exp_hic2 = exp_hic[cell_type_lower]
    *_,cbar = exp_hic1.plot_with(exp_hic2, fig=fig, ax=hic_ax, cbar_orientation=cbar_orientation)
    cbar.set_label('Interaction\nfrequency')

    # Median distances
    d1 = med_dists[cell_type_upper]
    d2 = med_dists[cell_type_lower]
    d1.plot_with(d2,fig=fig,ax=dist_ax, cbar_orientation=cbar_orientation,vmin=dist_vmin,vmax=dist_vmax)

    # Label cell types
    n = d1.num_beads
    for ax in [hic_ax,dist_ax]:
        ax.text(.995*n-1,.005*n,cell_type_upper,horizontalalignment='right',verticalalignment='top',
                color='black',path_effects=[pe.withStroke(linewidth=1, foreground="white")])
        
        ax.text(.005*n,.995*n-1,cell_type_lower,horizontalalignment='left',verticalalignment='baseline',
                color='black',path_effects=[pe.withStroke(linewidth=1, foreground="white")])

    # Remove the ylabels, which are in the DNase-seq plots
    hic_ax.set_ylabel('')
    dist_ax.set_ylabel('')

    return fig,(hic_ax,dist_ax)
```

## Plot

### DNase-seq data

```python
for chrom, start_indices in chrom_start_idx.items():
    for start_idx in start_indices:
        figs_axes = make_plots(chrom,start_idx)#, 'GM12878')

        for cell_type,(fig,ax) in figs_axes.items():
            fig.savefig(pdf_dir / f'part_a_dnase_chrom_{chrom}_{start_idx//1000}_{cell_type}.pdf',transparent=True)
```

### Distances, probabilities 

```python
for chrom,start_indices in chrom_start_idx.items():
    for k,si in enumerate(start_indices):
        fig, _ = plot_hic_distance(chrom,si,dist_vmin=dist_vmins[chrom][k],interp_nans = chrom=='chr21')
        fig.savefig(pdf_dir / f'part_a_hic_dists_chrom_{chrom}_{si//1000}.pdf',transparent=True)
```
