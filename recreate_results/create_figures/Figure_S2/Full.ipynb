## Prep

#### Import packages

```python
from ChromoGen import Conformations
from ChromoGen.data_utils import HiCDataset
from ChromoGen.model.Diffuser.training_utils import ConfigDataset
from pathlib import Path
from concurrent.futures import ThreadPoolExecutor
import torch
from tqdm.auto import tqdm
import matplotlib.pyplot as plt
plt.style.use('../support/plot_style.txt')
```

#### Filepaths

```python
cgen_conformations_dir = '../../downloaded_data/conformations/ChromoGen/genome_wide/GM12878/'
formatted_dipc_data = '../../downloaded_data/conformations/DipC/processed_data.h5'

mcool_fp = '../../outside_data/hic/GM12878_hg19.mcool'

pdf_dir = './pdfs/'
temp_data_dir = './temp_data/'

guidance_params = [(1.,0.),(5.,8.)]

resolution = 20_000

use_gpu = True
high_precision = True

max_workers = 8
```

#### Initialize objects

```python
cgen_conformations_dir = Path(cgen_conformations_dir)

pdf_dir = Path(pdf_dir)
pdf_dir.mkdir(exist_ok=True,parents=True)

temp_data_dir = Path(temp_data_dir)
temp_data_dir.mkdir(exist_ok=True,parents=True)

if use_gpu and torch.cuda.is_available():
    device = torch.device('cuda:0')
else:
    device = torch.device('cpu:0')

if high_precision:
    dtype = torch.double
else:
    dtype = torch.float

hic_dataset = HiCDataset(mcool_fp, resolution=resolution)

config_ds = ConfigDataset(
    formatted_dipc_data,
    segment_length=64,
    remove_diagonal=False,
    batch_size=0,
    normalize_distances=False, 
    geos=None,
    organisms=None,
    cell_types=None,
    cell_numbers=None,
    chroms=None, 
    replicates=None,
    shuffle=False,
    allow_overlap=True,
    two_channels=False,
    try_GPU=False,
    mean_dist_fp=None,
    mean_sq_dist_fp=None
)
```

## Analyze

#### Support

```python
def get_median_distance_maps_and_correlation(
    chrom, 
    start_idx,
    config_ds = config_ds,
    guidance_params=guidance_params,
    coord_dir=cgen_conformations_dir,
    hic_dataset=hic_dataset,
    device=device,
    dtype=dtype
):
    
    ####
    # Dip-C

    # Load conformations
    dipc_coords = Conformations(config_ds.fetch_specific_coords(chrom,start_idx)[-1]).to_(dtype=dtype,device=device)

    # Require minimum of 80 conformations to get a good median for Dip-C
    if len(dipc_coords) < 80:
        return None, None, float(torch.nan)
    
    # Compute median distance map
    dipc = dipc_coords.distances.median
    del dipc_coords
    
    ####
    # ChromoGen distances

    # Load conformations
    d = coord_dir
    cgen_coords = []
    for cs, rp in guidance_params:
        fn = f'chrom_{chrom}/start_idx_{start_idx//1000}_cond_scale_{cs}_rescaled_phi_{rp}.pt'
        cgen_coords.append(Conformations(d/fn, drop_invalid_conformations=True).to_(dtype=dtype,device=device))
    cgen_coords = cgen_coords[0].extend(cgen_coords[1:])

    # Compute median distance map
    cgen = cgen_coords.distances.median
    del cgen_coords

    ####
    # Compute the correlation coefficient between these
    n = cgen.num_beads
    i,j = torch.triu_indices(n,n,1,device=device)
    a = cgen.values[0,i,j]
    b = dipc.values[0,i,j]
    corr = (((a*b).mean() - a.mean() * b.mean()) / (a.std() * b.std())).cpu()
    
    #corr = float(torch.corrcoef(
    #    torch.stack([cgen.values[0,i,j], dipc.values[0,i,j]], dim=0)
    #)[0,1].cpu())

    # Place on CPU now that the computation is done to free up memory
    cgen.cpu_()
    dipc.cpu_()

    return cgen, dipc, corr
    
```

#### Compute all correlations

Ditch the median distance maps at this part since they're easy to load/create and take a lot of space if put in the save file

```python
save_file = temp_data_dir / 'correlations.pt'

if False:#save_file.exists():
    correlations = torch.load(save_file)
else:
    correlations = {}

    files = list(cgen_conformations_dir.rglob('*0.0.pt'))
    files.sort()

    with ThreadPoolExecutor(max_workers=max_workers) as executor:
        for f in files:
            chrom = f.parent.name.split('_')[-1]
            start_idx = int(f.name.split('_')[2]) * 1000
    
            if chrom not in correlations:
                correlations[chrom] = {}
            correlations[chrom][start_idx] = executor.submit(
                get_median_distance_maps_and_correlation,
                chrom,
                start_idx
            )

        for f in tqdm(files):
            chrom = f.parent.name.split('_')[-1]
            start_idx = int(f.name.split('_')[2]) * 1000
            correlations[chrom][start_idx] = correlations[chrom][start_idx].result()[-1]

        torch.save(correlations, save_file)
```

#### Select regions to plot (the best, median, and worst regions, per PCC)

Can easily sort a list of tuples based on the first index to keep track of region, etc., while identifying the regions to plot

```python
sorted_data = []

for chrom, sub_dict in correlations.items():
    for start_idx, corr in sub_dict.items():
        ####
        # drop NaN values
        if not corr > -2:
            continue
        ####
        sorted_data.append( (corr, chrom, start_idx) )

sorted_data.sort()
worst_regions = sorted_data[:3]
best_regions = list(reversed(sorted_data[-3:]))

# Using torch to make sure I don't screw the median part up 
sorted_data1 = torch.tensor([c for c,_,_ in sorted_data])
center_idx = torch.where(sorted_data1==sorted_data1.median())[0]

#center_idx = len(sorted_data)//2 - 1
#if len(sorted_data)%2 == 1:
#    center_idx-= 1

median_regions = list(reversed([sorted_data[center_idx-1], sorted_data[center_idx], sorted_data[center_idx+1]]))
```

## Plot

#### Support

```python
fig, axes = plt.subplots(ncols=6, nrows=3, figsize=(8,6), layout='constrained')

for i,regions in enumerate([worst_regions, median_regions, best_regions]):
    axes1 = axes[i]

    for j, (corr, chrom, start_idx) in enumerate(regions):

        ################
        # Load data
        cgen_dists, dipc_dists, _ = get_median_distance_maps_and_correlation(chrom,start_idx)
        stop_idx = start_idx + resolution * cgen_dists.num_beads
        exp_hic = hic_dataset.fetch(chrom, start_idx, stop_idx)

        # to nm
        cgen_dists._values*= 100
        dipc_dists._values*= 100

        ################
        # Plot

        ####
        # Experimental Hi-C
        ax = axes1[2*j]
        *_,cbar = exp_hic.plot(fig=fig,ax=ax,cbar_orientation='horizontal')
        if i < 2:
            cbar.set_label('')

        if j > 0:
            ax.set_ylabel('')

        s1 = round(start_idx/1e6, 2)
        s2 = round(stop_idx/1e6, 2)
        ax.set_title(f'Chromosome {chrom}:''\n'f'{s1}-{s2} Mb')

        ####
        # Median distances
        ax = axes1[2*j+1]

        *_, cbar = dipc_dists.plot_with(cgen_dists, fig=fig, ax=ax, cbar_orientation='horizontal')

        if j < 2:
            cbar.set_label('')

        ax.set_ylabel('')
        ax.set_title('$\\rho='f'{round(float(corr),4)}$')
        
fig.savefig(pdf_dir/'best_median_worst_correlations.pdf',transparent=True)
```
